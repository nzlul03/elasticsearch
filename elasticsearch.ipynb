### Search Query

#### 1. match "gempa"

```python
%%bash

curl -X GET 'localhost:9200/id_headlines/headline/_search?pretty' \
-H "Content-Type: application/json" -d '{
  "query": {
    "match": {
      "text": {
        "query": "gempa"
      }
    }
  }
}'
```

#### 2. match frase "bahan bakar"

```python
%%bash

curl -X GET 'localhost:9200/id_headlines/headline/_search?pretty' \
-H "Content-Type: application/json" -d '
{"query": 
  {"match_phrase": 
    {"text": "bahan bakar"}
  }
}
'
```

#### 3. boolean query "luka AND bakar"

```python
%%bash

# boolean query
# luka AND bakar

curl -X GET 'localhost:9200/id_headlines/headline/_search?pretty' -H 'Content-Type: application/json' -d '
{
  "query": {
    "bool": {
      "must": [
        {"term": {"text": "luka"}},
        {"term": {"text": "bakar"}}
      ]
    }
  }
}
'
```

#### 4. boolean query "presiden AND NOT jokowi"


```python
%%bash

# boolean query
# presiden AND NOT jokowi
curl -X GET 'localhost:9200/id_headlines/headline/_search?pretty' -H 'Content-Type: application/json' -d '
{
  "query": {
    "bool": {
      "must": 
        {"term": {"text": "presiden"}},
      "must_not": 
        {"term": {"text": "jokowi"}}
    }
  }
}'
```

#### 5. fuzzy query "padam"

```python
%%bash

curl -X GET 'localhost:9200/id_headlines/headline/_search?pretty' \
-H "Content-Type: application/json" -d '{
  "query": {
    "match": {
      "text": {
        "query": "padam",
        "fuzziness": "AUTO"
      }
    }
  }
}'
```

### Convert dataframe menjadi format SQuAD

```python
# Define list
paragraphs = []

# Variable initialization
paragraph = None
question = None
last_context = None
last_question = None
last_document_idx = 1000
last_question_idx = 2000
last_answer_idx = 3000
```

```python
from tqdm import tqdm
```

```python
%%bash

curl -X PUT "localhost:9200/idx_text"
```

```python
for idx, row in tqdm(train.iterrows()):
  if row['context'].strip() != last_context:
    last_document_idx += 1

    paragraph = {
        'context': row['context'],
        'document_idx': last_document_idx,
        'qas': []
    }

    paragraphs.append(paragraph)
    last_context = row['context'].strip()
  
  if row['question'].strip() != last_question:
    last_question_idx += 1

    question = {
        'id': last_question_idx,
        'question': row['question'].strip(),
        'is_impossible': False,
        'answers':[]
    }

    paragraph['qas'].append(question)
    last_question = row['question'].strip()
  
  last_answer_idx += 1

  answer = {
      "answer_id":last_answer_idx,
      "document_id": last_document_idx,
      "question_id": last_question_idx,
      "text": str(row['answer']).strip(),
      "answer_start": row['answer_start'],
      "answer_end": row['answer_start'] + len(str(row['answer']).strip()),
      "answer_category": None
  }
  question["answers"].append(answer)

reader_train = {
    'data' : [
        {
            'paragraphs': paragraphs
        }
    ]
}
```

## Initialize Retriever, Reader dan Pipeline

### Retriever

Retriever membantu membatasi cakupan pembaca ke unit text yang lebih kecil.

Berdasarkan soal Tugas 2, maka akan digunakan `DensePassageRetrieval` berdasarkan pada [Tutorial 6](https://colab.research.google.com/github/deepset-ai/haystack/blob/master/tutorials/Tutorial6_Better_Retrieval_via_DPR.ipynb).

```python
from haystack.nodes import DensePassageRetriever

retriever = DensePassageRetriever(
    document_store=document_store,
    query_embedding_model="facebook/dpr-question_encoder-single-nq-base",
    passage_embedding_model="facebook/dpr-ctx_encoder-single-nq-base",
    max_seq_len_query=64,
    max_seq_len_passage=256,
    batch_size=16,
    use_gpu=True,
    embed_title=True,
    use_fast_tokenizers=True,
)
document_store.update_embeddings(retriever)
```

### Reader

Untuk reader, akan digunakan FARMReader dengan the deepset/roberta-base-squad2 model (https://huggingface.co/deepset/roberta-base-squad2)

```python
reader = FARMReader(model_name_or_path="deepset/roberta-base-squad2", use_gpu=True)
```

```python
reader.train(data_dir=paragraph, train_filename="train", use_gpu=True, n_epochs=1, save_dir="my_model")
```

### Pipeline

Stick building blocks ke sebuah search pipeline.

Untuk mempercepat proses, Haystack dilengkapi dengan sebuah predefined Pipelines.

Salah satunya adalah `ExtractiveQAPipeline` yang mengkombinasikan sebuah retriever dan sebuah reader untuk menjawab pertanyaan.

```python
from haystack.pipelines import ExtractiveQAPipeline

pipe = ExtractiveQAPipeline(reader, retriever)
```

## Prediksi

Melakukan prediksi terhadap data testing.
* Retriever = top 5
* Reader = top 1

```python
prediction = pipe.run(
    query="", params={"Retriever": {"top_k": 5}, "Reader": {"top_k": 1}}
)
```
